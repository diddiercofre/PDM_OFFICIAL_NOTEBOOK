### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
# 3.1 Ruido Gaussiano

# Importamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,10,100) # Definimos un rango de valores para t, que representa el tiempo de 0 a 10 con 100 puntos
ruido = np.random.normal(0, 0.1, size=t.shape) # Generamos ruido gaussiano con media 0 y desviación estándar 0.1

funcion_normal = np.sin(t) # Definimos la función base sin ruido (Una onda seno)
funcion_con_ruido = funcion_normal + ruido # Sumamos el ruido gaussiano a la función seno para obtener la señal con ruido

# Graficamos ambas señales
plt.figure(figsize=(10,5))
plt.plot(t, funcion_normal, label="Señal sin ruido", linestyle="dashed", color="blue") # Graficamos la señal sin ruido (línea discontinua azul)
plt.plot(t, funcion_con_ruido, label="Señal con ruido", color="red", alpha=0.7) # Graficamos la señal con ruido (línea roja continua)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Señal con ruido gaussiano vs señal sin ruido") # Título de la gráfica
plt.legend()
plt.grid() # Agregamos cuadrícula
plt.show() # Mostramos la gráfica
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

```python
# EXPLICACIÓN:
# El ruido blanco gaussiano tiene una densidad espectral de potencia constante en todas las frecuencias, 
# mientras que el ruido rosa tiene una densidad espectral de potencia que disminuye a medida que las frecuencias 
# son mayores, especificamente a razón de 3 dB por octava. Esto significa que el ruido rosa tiene más energía 
# en las frecuencias bajas y menos en las frecuencias altas, mientras que en el ruido gaussiano es uniforme

# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Definir parámetros
fs = 1000  # Frecuencia de muestreo en Hz
N = 100000  # Número de muestras

# Generar ruido blanco gaussiano (media 0, varianza 1)
ruido_blanco = np.random.normal(0, 1, N)

# Generar ruido rosa usando filtrado en el dominio de la frecuencia
f = np.fft.rfftfreq(N, d=1/fs)
f[0] = 1  # Evitar división por cero
espectro_ruido_rosa = np.fft.rfft(ruido_blanco) / np.sqrt(f)
ruido_rosa = np.fft.irfft(espectro_ruido_rosa)

# Calcular densidad espectral de potencia (PSD)
f_w, Pxx_w = welch(ruido_blanco, fs, nperseg=1024)
f_p, Pxx_p = welch(ruido_rosa, fs, nperseg=1024)

# Graficar los espectros de potencia
plt.figure(figsize=(10,5))
plt.semilogy(f_w, Pxx_w, label="Ruido Blanco", color="blue")
plt.semilogy(f_p, Pxx_p, label="Ruido Rosa", color="red")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Densidad espectral de potencia")
plt.title("Comparación de Espectros: Ruido Blanco vs Ruido Rosa")
plt.legend()
plt.grid()
plt.show()

```
